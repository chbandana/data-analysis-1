```python
# 2)Can you tell me the Distribution of Vehicles based on Year of Registration with the help of a plot
```

```python
datac["yearOfRegistration"] = datac["yearOfRegistration"].where((datac["yearOfRegistration"] >= 1950) & (datac["yearOfRegistration"] <= 2023),datac["yearOfRegistration"].mode()[0])
plt.figure(figsize=(15, 7))
sns.histplot(data=datac, x=datac["yearOfRegistration"], edgecolor="b",linewidth=2)
plt.title("Distribution of vechicles in the year of Registration")

plt.show()
```

```python
# 3) Create a plot based on the Variation of the price range by the vehicle type
```

```python
datac['vehicleType'].value_counts()
```

```python
average_prices=datac.groupby('vehicleType')['price'].mean()
average_prices=average_prices.sort_values()
plt.figure(figsize=(10, 6))
sns.barplot(x=average_prices.index, y=average_prices.values)
plt.xlabel('Vehicle Type')
plt.ylabel('Average Price')
plt.title('Average Price by the Vehicle Type')

plt.show()
```
